# Remove Endcap Data / Barrel Data Remains

```python
events_P = []
for i in range(len(con)):
    
    #a = str(i)
    #thing = "binned_genPt_" + a
    if con['p'].iloc[i] > 0.00001:
        events_P.append(con.iloc[i])
        
events_P = pd.DataFrame(events_P)        
```

```python
events_P = events_P.set_axis(range(len(events_P)), axis='index') #set index ('one of these are unecessary')
```

# Correction factor 

```python
correction_factor= (events_P['Ccorrecal']+events_P['Ccorrhcal'])/(events_P['ecal']+events_P['hcal'])
correction_factor=pd.DataFrame(correction_factor) 
#correction_factor
```

```python
plt.hist(correction_factor,
         50,
         density=False,
         histtype='step',
         lw = 2 ,
         color='b',
         edgecolor='k',
         alpha=0.75)
plt.xlabel('.')
plt.ylabel('.')
plt.title('correction factor')
plt.show()
```
